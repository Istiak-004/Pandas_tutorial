# How to perform Merge Operation

```python
player = ["korim","rohim","jobar"]
point = [7,8,9]
title = ["game1","game2","game3"]
df1 = pd.DataFrame({"titles":title,"players":player,"points":point})
df1
```

```python
player = ["jon","bin","sam"]
point = [7,9,6]
title = ["game1","game2","game3"]
skill = ["punch","kick","jump"]
df2 =pd.DataFrame({"titles":title,"players":player,"points":point,"skills":skill})
df2
```

```python
player = ["kal","bon","kam"]
point = [7,9,6]
title = ["game3","game7","game9"]
skill = ["punch","kick","jump"]
df3 =pd.DataFrame({"titles":title,"players":player,"points":point,"skills":skill})
df3
```

```python
#inner Merge
df2.merge(df1,on=("titles","points"), how="inner") # here on = "on which title you need ", how = "how you wanna take" 
```

```python
#Inner Merge 
df2.merge(df3,on="skills",how="inner") 
```

```python
#Left merge
df1.merge(df3,on = "players",how = "left")
```

```python
#right merge
df1.merge(df3,on = "titles",how="right")
```

```python
#outer merge
df1.merge(df3,on = "titles",how = "outer")
```

# how to perform join operation in pandas

$ incase of joint two data point are joined on Index name

```python
player = ["kal","bon","kam"]
point = [7,9,6]
title = ["game3","game7","game9"]
skill = ["punch","kick","jump"]
df3 =pd.DataFrame({"Titles":title,"Players":player,"Points":point,"skills":skill},index=["L1","L2","L3"])
df3
```

```python
player = ["korim","rohim","jobar"]
point = [7,8,9]
title = ["game1","game2","game3"]
df1 = pd.DataFrame({"titles":title,"players":player,"points":point},index=["L4","L2","L1"])
df1
```

```python
#inner join
df1.join(df3,how="inner")
```

```python
#left join
df1.join(df3,how="left")
```

```python
#right join
df1.join(df3,how="right")
```
